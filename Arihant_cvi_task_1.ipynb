```
import cv2
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.lines as lines
import os
from google.colab import drive
```

```
image_path = 'drive/My Drive/OpenCV/Image.jpg'
image = cv2.imread(image_path)
plt.imshow(image)
```

```
#convert to gray scale
gray_img = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)
plt.imshow(gray_img, cmap= 'gray')
print(gray_img)
```

```
#Edge detection
edge_image = cv2.GaussianBlur(gray_img, (3,3), 1)
edge_image = cv2.Canny(edge_image, 30, 120)
plt.imshow(edge_image, cmap= 'gray')
```

```
edge_image = cv2.dilate(edge_image, cv2.getStructuringElement(cv2.MORPH_RECT, (5, 5)),iterations=1)
edge_image = cv2.erode( edge_image, cv2.getStructuringElement(cv2.MORPH_RECT, (5, 5)),iterations=1)
plt.imshow(edge_image, cmap = 'gray')
```
